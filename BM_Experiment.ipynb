```python
import pandas as pd

# Load the Excel file
df = pd.read_excel('clean_BM.xlsx', header=None)

sampled_entries = df[0].sample(n=200, random_state=1)  # `random_state` is set for reproducibility

for s in sampled_entries:
    print(s)


```

```python
# Convert the Series of sampled entries into a list
sampled_entries_list = sampled_entries.tolist()

# Now `sampled_entries_list` is a Python list
# You can print it or work with it as with any list
print(sampled_entries_list)

```

```python
import json

file_name = "bm_experiment_multishot.json"

def save_progress_and_responses(last_processed_index, responses):
    with open(file_name, 'w') as file:
        data = {
            'last_processed_index': last_processed_index,
            'responses': responses
        }
        json.dump(data, file)


def load_progress_and_responses():
    try:
        with open(file_name, 'r') as file:
            data = json.load(file)
            return data['last_processed_index'], data['responses']
    except FileNotFoundError:
        return -1, []  # No progress file, start from the beginning

    
import openai
  
api_key = ""
client = openai.OpenAI(api_key=api_key)


# Load the last processed index and existing responses
last_processed_index, responses = load_progress_and_responses()

datas = sampled_entries_list

# Loop through each user post starting from the last processed index
for i in range(last_processed_index + 1, len(datas)):
    post = datas[i]

    print(f"Processing post: {i + 1}/{len(datas)}")
    print(post)
    
    response = client.chat.completions.create(
        model="gpt-4-1106-preview",
        messages=[
            {
                "role": "system",
                "content": "You are an Adoption Agent who specializes in summarizing feedback notes from birth mothers examining a specific adoptive family profile. For the given note, answer: what are the birth mother’s main observations about the adoptive family? Example observations: Salary, Age, Occupation, Race, Religion, Order of presentation, timing, siblings, etc, or N/A. Example response: Age: older than self, Siblings: two sisters"
            },
            {
                "role": "user",
                "content": post[:50000]
            }
        ]
    )
    response_text = response.choices[0].message.content
    responses.append(response_text)

    # Save progress and responses every 10 posts
    if (i + 1) % 10 == 0:
        save_progress_and_responses(i, responses)
```
